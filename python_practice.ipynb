```python
def square_root(n):
    # Handle edge cases
    if n < 0:
        raise ValueError("Cannot calculate square root of negative number")
    if n == 0:
        return 0
    
    root = n / 2  # initial guess will be 1/2 of n
    for k in range(20):
        root = 0.5 * (root + (n / root))

    return root

# Convert input to number (float)
x = float(input("Enter your number: "))
y = square_root(x)

print(y)
```
