```python
#IF statements

#n = int(input("Enter any number :"))  # Convert input to integer
n = 6

if n % 2 == 0 :
    print("Even")
else:  # No condition needed here
    print("Odd")
```

```python
# Loops

for i in range(5):
    print(i)
```

```python
for i in range(2,6):
    print(i)
```

```python
for i in range(5,1,-1):
    print(i)
```

```python
#arrays

arr = [1,2,3]
print(arr)

#Can be used as a stack
arr.append(4)
arr.append(5)
arr.append(6)

print(arr)

## Then pop
arr.pop()
print(arr)


arr.insert(1,7)
print(arr)

arr[0] = 0
arr[3] = 0
print(arr)
```

```python
# Initialize arr of the size n with default value of 1

n = 5 
arr = [1]*5
print(arr)
print(len(arr))
```

```python
#sublisting(slicing)

arr = [1,2,3,4,5]
print(arr[1:3])
print(arr[0:5])
```

```python
# Loop through arrays

num = [1,2,3]

#Using index

for i in range(len(num)):
    print(num[i])

#Without index

for n in num:
    print(n)

```
